# t-SNE 範例
* 觀察S曲線使用 t-SNE 不同 perplexity 的流形還原效果

```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter
from sklearn import datasets
from time import time
```

```python
# 設定模型與繪圖參數
n_samples = 300
n_components = 2
perplexities = [4, 6, 9, 14, 21, 30, 45, 66, 100]

# 設定S曲線資料點
X, y = datasets.samples_generator.make_s_curve(n_samples, random_state=0)
```

# 作業
* 將原始資料集換為S型圖, 觀察不同 perplexity 下的流形還原結果

```python
from sklearn.manifold import TSNE

# 繪製不同 perplexity 下的 t-SNE 分群圖 (請注意! 語法也要對應調整)
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
ravel = axes.ravel()

ax = ravel[0]
ax.scatter(X[:, 0], X[:, 2], c=y)
ax.set_title('original')

for ax, perplexity in zip(ravel[1:], perplexities):
    tsne = TSNE(
        n_components=n_components,
        init='random',
        random_state=0,
        perplexity=perplexity
    )
    X_tmp = tsne.fit_transform(X)
    ax.scatter(X_tmp[:, 0], X_tmp[:, 1], c=y)
    ax.set_title('perplexity = {}'.format(perplexity))

plt.show()
```
